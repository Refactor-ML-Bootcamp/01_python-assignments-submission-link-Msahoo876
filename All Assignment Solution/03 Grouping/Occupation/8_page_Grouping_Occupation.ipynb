```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from this address.
# Step 3. Assign it to a variable called users.

import pandas as pd
import numpy as np
```

```python
user = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
user
```

```python
# Step 4. Discover what is the mean age per occupation

user.groupby('occupation').age.mean()
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least

user['Is_Male'] = user['gender'].apply(lambda x : True if x == 'M' else False)
(user.groupby('occupation').Is_Male.sum()/user.groupby('occupation').gender.count()).sort_values(ascending=True)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages

user.groupby('occupation').age.agg(['min','max'])
```

```python
# Step 7. For each combination of occupation and gender, calculate the mean age

user.groupby(['occupation','gender']).age.mean()
```

```python
# Step 8. For each occupation present the percentage of women and men

user_group1 = user.groupby(['occupation','gender']).agg({'gender':'count'})
user_group2 = user.groupby('occupation').count()
occ_group = user_group1.div(user_group2, level = 'occupation')
occ_group.loc[:,'gender']
```
